```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

movie_df = pd.read_excel("종합.xlsx")
y = movie_df["흥행등급"].values
```

```python
del movie_df["NAME"]
del movie_df["UBD"]
del movie_df["흥행등급"]
del movie_df["긍정댓글"]
del movie_df["개봉 9일째 관객수"]
del movie_df["관객증가율"]

X = movie_df.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
softmax_reg = LogisticRegression(multi_class='multinomial',solver='lbfgs',C=500,random_state=0)
softmax_reg.fit(X,y)
print("훈련 세트 정확도: {:.2f}".format(softmax_reg.score(X_train, y_train)))
print("테스트 세트 정확도: {:.2f}".format(softmax_reg.score(X_test, y_test)))
```

```python
print("훈련 세트 정확도: {:.2f}".format(softmax_reg.score(X_train, y_train)))
print("테스트 세트 정확도: {:.2f}".format(softmax_reg.score(X_test, y_test)))
```

```python
C_settings = range(1,1000)

training_accuracy = []
test_accuracy = []
for C_setting in C_settings:
    # 모델 생성
    softmax_reg = LogisticRegression(multi_class='multinomial',solver='lbfgs',C=C_setting,random_state=0)
    softmax_reg.fit(X_train, y_train)
    # 훈련 세트 정확도 저장
    training_accuracy.append(softmax_reg.score(X_train, y_train))
    # 일반화 정확도 저장
    test_accuracy.append(softmax_reg.score(X_test, y_test))
    
plt.plot(C_settings, training_accuracy, label="train")
plt.plot(C_settings, test_accuracy, label="test")
plt.ylabel("accuracy")
plt.xlabel("C_settingss")
plt.legend()
plt.show()
```
