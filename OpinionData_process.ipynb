```python
movie_name = open('./영화제목.txt',encoding = 'utf8')
movie_title = [num for num in movie_name.read().split('\n')]
len(movie_title)
```

```python
positive_list = []
for i in range(len(movie_title)):
    data= pd.read_excel('positive_negative/'+movie_title[i]+'.xlsx')
    positive_list.append(data['label'].value_counts()[1])
positive_list
```

```python
series = pd.Series(positive_list,name='댓글긍정')
```

```python
data = data.join(series)
```
